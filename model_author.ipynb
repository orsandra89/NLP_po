```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Poetry_Data_with_Epochs.csv")


author_counts = df["author"].value_counts()
valid_authors = author_counts[author_counts > 1].index
df_filtered = df[df["author"].isin(valid_authors)].copy()


X = df_filtered.drop(columns=["filename", "author", "epoch"]).fillna(0)
y = df_filtered["author"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)
print("ðŸ‘¤ Classification Report (Authors):")
print(classification_report(y_test, y_pred))

```
